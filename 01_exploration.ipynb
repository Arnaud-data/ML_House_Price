```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Imports OK")
```

```python
df = pd.read_csv(r"C:\Users\arnau\Documents\Kaggle_House_Price\Kaggle_House_Price_DATA\train.csv")
print(f"Shape : {df.shape}")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
sns.histplot(df['SalePrice'], kde=True)
plt.title("SalePrice Distribution")
plt.show()
```

```python
skewness = df['SalePrice'].skew()
print(f"Skewness : {skewness}")
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

# Visualisation
sns.barplot(x=missing.values, y=missing.index)
plt.title("Valeurs manquantes par colonne")
plt.show()
```

```python
corr = df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
print(corr.head(15))

top_features = corr.head(11).index
sns.heatmap(df[top_features].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

```python
cat_cols = df.select_dtypes(include='str').columns

for col in cat_cols:
    sns.boxplot(x=df[col], y=df['SalePrice'])
    plt.xticks(rotation=45)
    plt.title(f"SalePrice par {col}")
    plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

top_num = ['GrLivArea', 'TotalBsmtSF', 'GarageArea',
           'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']

for ax, col in zip(axes.flatten(), top_num):
    ax.scatter(df[col], df['SalePrice'], alpha=0.5)
    ax.set_xlabel(col)
    ax.set_ylabel('SalePrice')

plt.tight_layout()
plt.show()
```
